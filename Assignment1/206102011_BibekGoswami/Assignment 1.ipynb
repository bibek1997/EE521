1. Continuous Signal

Basic Signal

```python
def u(t):
    x = []
    for i in range(len(t)):
        if t[i] < 0:
            x.append(0)
        else:
            x.append(1)
    return x

def r(t):
    x = []
    for i in range(len(t)):
        if t[i] < 0:
            x.append(0)
        else:
            x.append(t[i])
    return x

def p(t):
    x = []
    for i in range(len(t)):
        if t[i] < 0:
            x.append(0)
        else:
            x.append((t[i]**2)/2)
    return x

def p(t):
    x = []
    for i in range(len(t)):
        if t[i] < 0:
            x.append(0)
        else:
            x.append((t[i]**2)/2)
    return x

def rect(t,tou):
    x = []
    for i in range(len(t)):
        if t[i] < (-tou/2) or t[i] > (tou/2):
            x.append(0)
        else:
            x.append(1)
    return x

def tri(t,tou):
    x = []
    for i in range(len(t)):
        if t[i] < (-tou/2) or t[i] > (tou/2):
            x.append(0)
        elif t[i] > (-tou/2) and t[i] < 0:
            x.append(1+((2/tou)*t[i]))
        elif t[i] > 0 and t[i] < (tou/2):
            x.append(1-(2*t[i]/tou))
    return x

def sinc(t):
    x = []
    for i in range(len(t)):
        x.append((np.sin(np.pi * t[i]) / (np.pi * t[i])))
    return x

t = np.linspace(-10,10,100)
```

```python
# Aexp(Bt)cos(wt)
A = np.array([0.1,0.5,1,2])
B = np.array([-0.25,-0.5,-1,0.5,1])
w = np.array([2 * np.pi, np.pi / 6, 5/3 * np.pi])

for i in range(A.shape[0]):
    for j in range(B.shape[0]):
        for k in range(w.shape[0]):
            y = A[i] * np.exp(B[j] * t) * np.cos(w[k] * t)
            plt.plot(t,y)
            
plt.show()
```

```python
#Aexp(j(w*t+theta))
def y(A,w,theta,t):
    im = A * np.cos((w*t)+theta)
    re = A * np.sin((w*t)+theta)
    
    return im,re
```

```python
#(A,w,theta) = (1,0.14*pi,0)
y_im, y_re = y(1, 0.14*np.pi, 0,t)

fig = plt.figure()

plt.subplot(1,2,1)
plt.plot(t,y_re)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Re(y(t))')

plt.subplot(1,2,2)
plt.plot(t,y_im)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Im(y(t))')

plt.show()
```

2. Discrete Signal

```python
def delta_fun(n):
    x=[]
    for i in range(len(n)):
        if(n[i]==0):
            x.append(1)
        else:
            x.append(0)
            
    return x
    
n = np.arange(-10,11)
```

```python
x1_n = delta_fun(n)

plt.stem(n,x1_n)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Delta of n')
```

3. Systems (Time domain approach)

```python
def convolution(inputSeq, impulseResponse):
    outputSeq = np.convolve(inputSeq,impulseResponse)
    
    return outputSeq
```

```python
x_n = np.array([2,4,5,2,7])
h_n = np.array([8,-5,4])

y_n = convolution(x_n,h_n)
print(y_n)
```

```python
h_n = 2*np.array(delta_fun(n-3))
n1 = np.arange(-20,21)
```

```python
u_n = u(n)
y_unit_step = convolution(u_n,h_n)

plt.stem(n1,y_unit_step)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Response of unit step function')
```

```python
r_n = r(n)
y_ramp = convolution(r_n,h_n)

plt.stem(n1,y_ramp)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Response of ramp function')
```

```python
p_n = p(n)
y_parabolic = convolution(p_n,h_n)

plt.stem(n1,y_parabolic)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Response of parabolic function')
```

```python
re_complex_exp = np.exp(2*n) * np.cos(3*n)
im_complex_exp = np.exp(2*n) * np.cos(3*n)

y_re_complex_exp = convolution(re_complex_exp,h_n)
y_im_complex_exp = convolution(im_complex_exp,h_n)

fig = plt.figure(figsize=(8,10))

plt.subplot(2,1,1)
plt.stem(n1,y_re_complex_exp)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Response of real part complex exponential')

plt.subplot(2,1,2)
plt.stem(n1,y_im_complex_exp)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Response of imaginary part complex exponential')

plt.show()
```
